# Exploratory Data Analysis

# Univariate analysis

```python

for i in df[['delivery_person_age','time_taken','delivery_person_ratings']]:
    sns.histplot(x = i,kde=True,data=df)
    plt.show()
```

### It shows how the data is distributed for every single column like an age distribution of delivery persons, 
###  time taken for delivery, ratings distribution.

```python
for i in df[['weather_conditions','road_traffic_density', 
             'type_of_order','festival', 'city']]:

    ax=sns.countplot(x=i, hue=i,data=df)
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='baseline', fontsize=11, color='black')
    plt.title(i)
    plt.show()
```


# Bivariate analysis

```python
sns.boxplot(x='road_traffic_density',
            y='time_taken',
            hue='road_traffic_density',
            data=df
            )
plt.title("Delivery Time by Traffic Density")
plt.show()

sns.boxplot(x='weather_conditions',
            y='time_taken',
            hue='weather_conditions',
            data=df            
)
plt.title("Delivery Time by weather conditions")
plt.show()

sns.boxplot(x='multiple_deliveries',
            y='time_taken',
            hue='multiple_deliveries',
            data=df            
)
plt.title("Delivery Time by multiple deliveries")
plt.show()
```

```python

sns.barplot(x='city',
            y='time_taken',
            hue='road_traffic_density',
            data=df
           )
plt.title("Delivery Time by City and Traffic Density")
plt.show()
```
